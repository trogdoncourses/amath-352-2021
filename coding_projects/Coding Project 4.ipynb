### Polynomial integration

Consider computing
$$
 \int_{-1}^x P(y) dy, \quad P(y) = a_0 + a_1 y + a_2 y^2 + \cdots a_n y^{n}.
$$
This is something we can do.  For example,
$$
 \int_{-1}^x (a_0 + a_1 y + a_2 y^2) dy = \left.\left( a_0 y + a_1 \frac{y^2}{2} + a_2 \frac{y^3}{3} \right) \right|_{-1}^x\\
  = a_0 x + a_1 \frac{x^2}{2} + a_2 \frac{x^3}{3} - \left( -a_0 + \frac{a_1}{2} - \frac{a_2}{3} \right).
$$
We can keep track of a polynomial using vectors:
$$
 a_0 + a_1 x + a_2 x^2 \longrightarrow \begin{pmatrix} a_0 \\ a_1 \\ a_2 \end{pmatrix}
$$
and
$$
a_0 x + a_1 \frac{x^2}{2} + a_2 \frac{x^3}{3} - \left( -a_0 + \frac{a_1}{2} - \frac{a_2}{3} \right) \longrightarrow \begin{pmatrix} \displaystyle a_0 - \frac{a_1}{2} + \frac{a_2}{3} \\ a_0 \\\displaystyle\frac{a_1}{2} \\ \displaystyle\frac{a_2}{3} \end{pmatrix}
$$
The matrix that then performs this integration is:
$$
\begin{pmatrix} \displaystyle a_0 - \frac{a_1}{2} + \frac{a_2}{3} \\ a_0 \\\displaystyle\frac{a_1}{2} \\ \displaystyle\frac{a_2}{3} \end{pmatrix} = \begin{pmatrix} 1 & - \frac{1}{2} & \frac{1}{3} \\ 1 & 0 & 0 \\ 0 & \displaystyle \frac 1 2 & 0 \\ 0 & 0 & \displaystyle \frac 1 3 \end{pmatrix} \begin{pmatrix} a_0 \\ a_1 \\ a_2 \end{pmatrix}
$$
So, if you start with a quadratic polynomial, you'll have a $4 \times 3$ matrix.  If you start with a cubic, you'll have a $5 \times 4$ matrix.  In general, if you have a $k$th-order polynomial, you'll have a $(k+2) \times (k+1)$ matrix, call it $M_k$.  Consider the polynomial:
$$
  p(x) = 1 - \frac{x^2}{1!} + \frac{x^4}{2!} - \frac{x^6}{3!} + \cdots + \frac{x^{40}}{20!} = \sum_{j=0}^{20}(-1)^j \frac{x^{2j}}{j!}.
$$
Note that the sign changes between each term and this is a 40th-order polynomial.  Save $M_k$ to the variable `T3_1` and use it to compute
$$
   \int_{-1}^1 p(y) dy.
$$
Save this number to the variable `T3_2`
